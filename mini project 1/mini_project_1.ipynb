```python
# Cell 2: Define Arabic Letters and Diacritics
# Arabic Alphabet
Alef='Ш§'
Alef_Hamza = 'ШЈ'
Alef_Mad='Шў'
Hamza = 'ШЎ'
Beh='ШЁ'
Teh='ШЄ'
Theh='Ш«'
Jeem='Ш¬'
Hah='Ш­'
Khah='Ш®'
Dal='ШЇ'
Thal='Ш°'
Reh='Ш±'
Zain='ШІ'
Seen='Ші'
Sheen='Шґ'
Sad='Шµ'
Dad='Ш¶'
Tah='Ш·'
Zah='Шё'
Ain='Ш№'
Ghain='Шє'
Feh='ЩЃ'
Qaf='Щ‚'
Kaf='Щѓ'
Lam='Щ„'
Mem='Щ…'
Noon='Щ†'
Heh='Щ‡'
Wow='Щ€'
Yeh='ЩЉ'

# Arabic Diacritics (Tashkeel)
Fatha='ЩЋ'
Fatha_2='b'
Fatha_3='c'
Fatha_4='d'
Fatha_5='e'
Fatha_6='f'
Fatha_7='g'
Damma='ЩЏ'
Damma_2='='
Kasra='Щђ'
Kasra_2='-'
Shadda='Щ‘'
Sukun='Щ’'
double_fatha='Щ‹'
double_damma='ЩЊ'
double_kasra='ЩЌ'
```

```python
# Cell 3: Define Letter to Phoneme Mapping
letter_to_phoneme = {
    'Ш§':'/A/','ШЈ':'E','Шў':'/AE:/','ШЎ':'/E/','ШЁ':'/B/','ШЄ':'/T/',
    'Ш«':'/TH/','Ш¬':'/JH/','Ш­':'/HH/','Ш®':'/KH/','ШЇ':'/D/',
    'Ш°':'/DH/', 'Ш±':'/R/','ШІ':'/Z/','Ші':'/S/','Шґ':'/SH/',
    'Шµ':'/SS/','Ш¶':'/DD/','Ш·':'/TT/','Шё':'/DH2/','Ш№':'/AI/',
    'Шє':'/GH/','ЩЃ':'/F/','Щ‚':'/Q/','Щѓ':'/K/','Щ„':'/L/','Щ…':'/M/',
    'Щ†':'/N/', 'Щ‡':'/H/','Щ€':'/W/','ЩЉ':'/Y/','Щ‰':'/A/', 'ЩЋ':'/AE/',
    'ЩЏ':'/UH/', 'Щђ':'/IH/', 'a':'/AE:/', 'b':'/AH/',  'c':'/AA/',
    'd':'/AA:/','e':'/AH:/', 'f':'/AW/', 'g':'/AY/', 'Щ’':'',
    'Ш¦':'EIH', 'Ш¤':'EUH','ШҐ':'E', 'Ш©':'/T/','=':'/UX/','-':'/IX/'
}
```

```python
# Cell 4: Define Character Classes
L = 'ШЁШЄШ«Ш¬Ш­Ш®ШЇШ°Ш±ШІШіШґШµШ¶Ш·ШёШ№ШєЩЃЩ‚ЩѓЩ„Щ…Щ†Щ‡'   # All Arabic consonants
D = 'Щ‹ЩЊЩЌЩЋЩЏЩђЩ‘Щ’'                       # Diacritic marks
S = 'BOS'                            # Word start
T = 'EOS'                            # Word end
SH = 'ШЄШ«ШЇШ°Ш±ШІШіШґШµШ¶Ш·ШёЩ„Щ†'                # Shamsi Letters
V = 'ЩЋЩЏЩђЩ‘'                           # Vowels
VA = 'ЩЋЩЏЩђ'                           # Vowels without Shadda
P = 'Щ€ШЁЩЃЩѓЩ„'                          # Prefix letters
E = 'Ш·ШµШ¶Шё'                           # Emphatic letters
PH = 'Щ‚ШєШ®Ш±'                          # Pharyngeal letters
```

```python
# Cell 5: Define Rule to Regex Function (from HW2)
def rule_to_regex(word):
    # Group_1: Vowels
    word = re.sub(Fatha+'('+Wow+'(['+L+']|['+T+"]))",Fatha_6+r'\1',word)
    word = re.sub('(['+E+'])'+Fatha+Alef,r'\1'+Fatha_5,word)
    word = re.sub('(['+PH+'])'+Fatha+Alef,r'\1'+Fatha_3,word)
    word = re.sub(Fatha+'('+Yeh+'(['+L+']|['+T+"]))",Fatha_7+r'\1',word)
    word = re.sub('([^'+E+']|['+PH+'])'+Fatha+Alef,r'\1'+Alef_Mad,word)
    word = re.sub('([^'+E+']|['+PH+'])'+Fatha+'([^'+Alef+']|(('+Wow+'|'+Yeh+')(['+L+']|['+T+"])))",r'\1'+Fatha+r'\2',word)
    word = re.sub('(['+PH+'])'+Fatha+'([^'+Alef+']|(('+Wow+'|'+Yeh+')(['+L+']|['+T+"])))",r'\1'+Fatha_3+r'\2',word)
    word = re.sub('(['+E+'])'+Fatha+'([^'+Alef+']|(('+Wow+'|'+Yeh+')(['+L+']|['+T+"])))",r'\1'+Fatha_2+r'\2',word)

    # Group_2: Tanween
    word = re.sub('([^'+E+']|['+PH+'])'+double_fatha,r'\1'+Fatha+Noon,word)
    word = re.sub('(['+E+'])'+double_fatha,r'\1'+Fatha_2+Noon,word)
    word = re.sub('(['+PH+'])'+double_fatha,r'\1'+Fatha_3+Noon,word)
    word = re.sub('([^'+PH+'])'+double_damma,r'\1'+Damma+Noon,word)
    word = re.sub('(['+PH+'])'+double_damma,r'\1'+Damma_2+Noon,word)
    word = re.sub('([^'+PH+'])'+double_kasra,r'\1'+Kasra+Noon,word)
    word = re.sub('(['+PH+'])'+double_kasra,r'\1'+Kasra_2+Noon,word)
    word = re.sub('(.)'+Shadda+'(.)',r'\1\1\2',word)

    # Group_3: Semi Vowels
    word = re.sub('('+Fatha+'|'+Damma+')'+Yeh+'([^'+V+'])',r'\1\2',word)
    word = re.sub('('+Fatha+'|'+Kasra+')'+Wow+'([^'+V+'])',r'\1\2',word)

    # Group_4: Letter Lam
    word = re.sub('((['+P+V+'])?'+Alef+'?)'+Lam+'(['+SH+'])',r'\1\2\3',word)

    # Group_5: Consonants
    word = re.sub(Noon+Beh,Mem+Mem,word)
    word = re.sub(Dal+'('+Teh+'['+V+'])',r'\1',word)
    word = re.sub(Dad+'('+Teh+'['+V+'])',r'\1',word)

    return word
```

```python
# Cell 6: Convert Text to Phonemes
def text_to_phonemes(text):
    """Convert Arabic text to phoneme sequence"""
    words = text.split(' ')
    all_phonemes = []

    for word in words:
        # Clean word from verse numbers
        word = re.sub(r'пґї\d+пґѕ', '', word).strip()
        if not word:
            continue

        # Add BOS and EOS tags
        tagged_word = 'BOS' + word + 'EOS'

        # Apply regex rules
        mapped_word = rule_to_regex(tagged_word)

        # Convert to phonemes
        phoneme_seq = []
        for char in mapped_word:
            if char in letter_to_phoneme:
                phoneme = letter_to_phoneme[char]
                if phoneme:  # Skip empty strings
                    phoneme_seq.append(phoneme)

        all_phonemes.extend(phoneme_seq)

    return all_phonemes
```

```python
# Cell 7: Define Consonants and Vowels
CONSONANTS = {
    '/B/', '/T/', '/TH/', '/JH/', '/HH/', '/KH/', '/D/', '/DH/',
    '/R/', '/Z/', '/S/', '/SH/', '/SS/', '/DD/', '/TT/', '/DH2/',
    '/AI/', '/GH/', '/F/', '/Q/', '/K/', '/L/', '/M/', '/N/',
    '/H/', '/W/', '/Y/', 'E', 'EIH', 'EUH'
}

VOWELS = {
    '/AE/', '/AH/', '/AA/', '/AA:/', '/AH:/', '/AW/', '/AY/',
    '/UH/', '/UX/', '/IH/', '/IX/', '/AE:/', '/A/'
}
```

```python
# Cell 8: Phonemes to Syllable Patterns
def phonemes_to_syllables(phonemes):
    """Convert phoneme sequence to syllable patterns"""
    syllables = []
    i = 0

    while i < len(phonemes):
        phoneme = phonemes[i]

        # Skip BOS/EOS markers
        if phoneme in ['BOS', 'EOS']:
            i += 1
            continue

        # Check if it's a consonant
        if phoneme in CONSONANTS:
            syllable = 'C'
            i += 1

            # Look for vowels after consonant
            vowel_count = 0
            while i < len(phonemes) and phonemes[i] in VOWELS:
                syllable += 'V'
                vowel_count += 1
                i += 1

            # Look for consonants after vowel(s)
            consonant_count = 0
            while i < len(phonemes) and phonemes[i] in CONSONANTS and consonant_count < 2:
                # Check if next is vowel (start of new syllable)
                if i + 1 < len(phonemes) and phonemes[i + 1] in VOWELS:
                    break
                syllable += 'C'
                consonant_count += 1
                i += 1

            # Classify syllable
            if syllable in ['CV', 'CVC', 'CVV', 'CVVC', 'CVCC']:
                syllables.append(syllable)
            else:
                syllables.append('CX')  # Unknown pattern
        else:
            # Skip orphan vowels or unknown patterns
            syllables.append('CX')
            i += 1

    return syllables
```

```python
# Cell 9: Syllables to Numbers
def syllables_to_numbers(syllables):
    """Convert syllable patterns to number sequence"""
    mapping = {
        'CV': 1,
        'CVC': 2,
        'CVV': 3,
        'CVVC': 4,
        'CVCC': 5,
        'CX': 6
    }
    return [mapping[s] for s in syllables]
```

```python
# Cell 10: Main Processing Function
def process_quran_text(text, sura_name):
    """Process Quranic text and output syllable sequences"""
    print(f"\n{'='*60}")
    print(f"Processing: {sura_name}")
    print(f"{'='*60}")

    # Get phonemes
    phonemes = text_to_phonemes(text)
    print(f"\nPhonemes: {' '.join(phonemes[:30])}...")  # Show first 30

    # Get syllables
    syllables = phonemes_to_syllables(phonemes)
    print(f"\nSyllable Patterns: {' '.join(syllables)}")

    # Get numbers
    numbers = syllables_to_numbers(syllables)
    print(f"\nNumber Sequence: {' '.join(map(str, numbers))}")

    return {
        'sura': sura_name,
        'phonemes': phonemes,
        'syllables': syllables,
        'numbers': numbers
    }
```
